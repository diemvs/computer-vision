# 4. Работа с каналами

## CV2

```python
image_cv = cv2.imread('example.jpg')

blue, green, red = cv2.split(image_cv)

cv2.imwrite('results/4/blue_chill_guy.png', blue)
cv2.imwrite('results/4/green_chill_guy.png', green)
cv2.imwrite('results/4/red_chill_guy.png', red)

show(blue)
```

```python
merged_image = cv2.merge(( red, green, blue))

cv2.imwrite('results/4/merged_image.png', merged_image)

show(merged_image)
```

## PIL

```python
image_pil = Image.open('example.jpg')

r, g, b = image_pil.split()

r.save('results/4/red_chill_guy.jpeg')
g.save('results/4/green_chill_guy.jpeg')
b.save('results/4/blue_chill_guy.jpeg')
```

```python
merged_image = Image.merge('RGB', (r, g, b))

merged_image.save('results/4/merged_image.jpeg')

show(merged_image)
```

# 5. Измерение размера и обрезка изображений

## CV2

```python
image_cv = cv2.imread('example.jpg')

resized_cv = cv2.resize(image_cv, (100, 100))

cv2.imwrite('results/5/resized_chill_guy.png', resized_cv)

height, width = resized_cv.shape[:2]

start_x = width//2 - 50
start_y = height//2 - 10

cropped_cv = resized_cv[start_y:start_y+200, start_x:start_x+200]

cv2.imwrite('results/5/cropped_chill_guy.png', cropped_cv)
```

```python
show(cropped_cv)
```

```python
show(resized_cv)
```

## PIL

```python
image_pil = Image.open('example.jpg')

resized_pil = image_pil.resize((100, 100))
resized_pil.save('results/6/resized_pil.png')

# Обрезка центральной области
width, height = resized_pil.size
left = width //2 - 50
upper = height //2 - 50
right = width /2 + 50
lower = height //2 + 50

cropped_pil = resized_pil.crop((left, upper, right, lower))
cropped_pil.save('results/6/cropped_pil.png')
```

```python
show(resized_pil)
```

```python
show(cropped_pil)
```

## 7. Поворот изображения на произвольный угол

## CV2

```python
image_cv = cv2.imread('example.jpg')

height, width = image_cv.shape[:2]

center = (width // 2, height // 2)

matrix_clockwise = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated_clockwise_cv = cv2.warpAffine(image_cv, matrix_clockwise, (image_cv.shape[1], image_cv.shape[0]))
cv2.imwrite('results/7/rotated_chill_guy.png', rotated_clockwise_cv)
```

```python
show(rotated_clockwise_cv)
```

## PIL

```python
image_pil = Image.open('example.jpg')

rotated_clockwise_pil = image_pil.rotate(-45, expand=True)
rotated_clockwise_pil.save('results/7/rotated_clockwise_pil.png')
```

```python
show(rotated_clockwise_pil)
```

# 8. Гистограммаизображения

## CV2

```python
import cv2
import matplotlib.pyplot as plt

image_cv = cv2.imread('example.jpg')

hist = cv2.calcHist([image_cv], [1], None, [256], [0, 256])

plt.figure(figsize=(10,5))
plt.plot(hist, color='black')
plt.title('Гистограмма зеленого канала')
plt.xlabel('Интенсивность пикселей')
plt.ylabel('Частота')
plt.show()
```
